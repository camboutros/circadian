```python
####### Statistics formulas coded in python, provided opensource by malishoaib at https://malishoaib.wordpress.com/about/ ###
 
def newsum(arg1, arg2 = 0, arg3 = False):
    # Arguments 2 and 3 are defaulted so this function can bu used as inbuilt sum() function.
    
    arg1 = np.nan_to_num(arg1)
    
    if arg3 == False:
        arg3 = len(arg1)
 
    if arg3 == arg2:
        return arg1[arg2]
 
    n = [] # initializing the list. I have written this function to store variables in a list and then indexs to initialize sum from and to teh desired region of the list.
    j = arg2
    for i in range(arg3 - arg2):
        n.append(arg1[j])
        j = j + 1
    return sum(n) # returning the sum() on list.

 
def cov(arg1, arg2):
 
    ##Formula for covariance is = 1/1-N . Sum (x_i - X)(y_i - Y) ___ OR ___ [Sum (x_i - X)(y_i - Y)] / N-1 ##
    arg1 = np.nan_to_num(arg1)
    arg2 = np.nan_to_num(arg2)
    
    # error messages...
    if len(arg1) != len(arg2):
        print('Error: Lengths of variables to be correlated must be equal')
        
    #Calculating X and Y = means
    mean_x = mean(arg1)
    mean_y = mean(arg2)
 
    # Calculating (x_i - X)... for i in vector 'x'
    mu = []
    n = 0
    for i in arg1:
        var_a = arg1[n] - mean_x
        mu.append(var_a)
        n = n + 1
 
    # Calculating (y_i - Y)... for i in vector 'y'
    nu = []
    n = 0
    for i in arg2:
        var_b = arg2[n] - mean_y
        nu.append(var_b)
        n = n + 1
 
    # Calculating the product --> (x_i - X)(y_i - Y)...
    n = 0
    lst = []
    for i in nu:
        var4 = nu[n] * mu[n]
        lst.append(var4)
        n = n + 1
    # Sum(x_i - X)(y_i - Y)
    var5 = sum(lst)
 
    # Finalizing the formula
    cov = var5 / ((len(arg1)) - 1)
    return cov
 
     
def variance(arg1):
    ## Formula for Variance is = 1/N-1 . Sum (x_i - X)^2 ___ OR ___ [Sum (x_i - X)^2] / N-1 ##
    arg1 = np.nan_to_num(arg1)
    # Calculating 'X' a.k.a 'mean' of the sample...
    var2 = mean(arg1)
 
  # Calculating (x-i - X)^2 ..
    n = 0
    lst = []
    for i in arg1[:]:
        var = arg1[n] - var2
        n = n + 1
        lst.append(var**2)
 
  # Calculating final expression ...
    variance = sum(lst) / (len(arg1) - 1)
 
  # Returning the value of function..
    return variance
#########################################################################
 
def sd(arg1):
    sd = math.sqrt(variance(arg1))
    return sd
 
#####################################################################
#######################################################################
 
def corr(arg1, arg2, df = False, pval = False):
    arg1 = np.nan_to_num(arg1)
    arg2 = np.nan_to_num(arg2)
    ## takes the formula = cov(x,y)/sd(x).sd(y)....
 
    var1 = cov(arg1, arg2)
    sd_x = sd(arg1)
    sd_y = sd(arg2)
    correlation = var1 / (sd_x * sd_y)
 
    #return abs(correlation)
    return correlation
 
##########################################################################

##########TEST


def autocorr(x, t=1):
    numpy.corrcoef(numpy.array([x[0:len(x)-t], x[t:len(x)]]))
```

```python
#####################################################################
# Additional methods 

def mean (arg1):
    arg1 = np.nan_to_num(arg1)
    return (sum(arg1)/len(arg1))

def normalize (arg1, min1, max1):
    norm = ((arg1 - min1) / (max1 - min1))
    return norm

#####################################################################
```

```python
############ PrettyMIDI Setup ##################################################
# Create a PrettyMIDI object          

temperature_sonify = pretty_midi.PrettyMIDI()

# Create an Instrument instance for a cello instrument. Choice of cello is 
# arbitrary; just needed the name of an instrument which exists in the 
# general MIDI standard to get a general MIDI program number

temperature_program = pretty_midi.instrument_name_to_program('Cello')

cello = pretty_midi.Instrument(program = temperature_program) # core
piano = pretty_midi.Instrument(program = temperature_program)  # distal
synth = pretty_midi.Instrument(program = temperature_program) # axial

flute = pretty_midi.Instrument(program = temperature_program) # diff_core
clarinet = pretty_midi.Instrument(program = temperature_program) # diff_distal
piccolo = pretty_midi.Instrument(program = temperature_program) # diff_axial

percussion = pretty_midi.Instrument(program = temperature_program) # max and min  

#                                                                                 #
###################################################################################

###################################################################################
#             Other Useful Variables                                              #

new_data_core_csv = np.nan_to_num(data_core)
new_data_distal_csv = np.nan_to_num(data_distal)
new_data_axial_csv = np.nan_to_num(data_axial)

new_diff_axial_csv = np.nan_to_num(diff_axial)
new_diff_core_csv = np.nan_to_num(diff_core)
new_diff_distal_csv = np.nan_to_num(diff_distal)

global_data_min = axial_min
global_data_max = core_max

global_diff_min = diff_axial_min
global_diff_max = diff_axial_max

reset = 0
time_shift = .1 # tenth of a second

octave_ratio = 4
floor_frequency = 220 #440 # 'A4', MIDI note 69
ceiling_frequency = 880 #1760 # 'A6', MIDI note 1760
maxpitchbend = 8091
velocity_value = 10

percussion_flag = False

#                                                                                 #                                                                   
###################################################################################

###################################################################################
#                      Conversion Methods                                         #

        
def percuss(celsius):
    if celsius not in (core_max, distal_max, axial_max, core_min, distal_min, axial_min):
        percussion_flag = False
        return percussion_flag
    else:
        percussion_flag = True
        return percussion_flag
        
def convert_data_to_midi (data_array, pretty_midi_instrument):
    
    time = reset 
    
    for celsius in data_array[:, 1]:
        
        #if celsius == core_max:
        #    celsius = (celsius * 30) 
        #    note_num = pretty_midi.hz_to_note_number(celsius)
        #    note_num = int(note_num)

        #    # print 'Percussion: Midi note number = ', note_num
        #    note = pretty_midi.Note(velocity = 100, pitch = note_num, start = time, end = time + .1)
        #    percussion.notes.append(note)
            
        if celsius == 0.0:
            celsius = global_data_min
        # print 'Celsius:', celsius
        percussion_flag = percuss(celsius)
        celsius = normalize(celsius, global_data_min, global_data_max) # convert to 0.0 - 1.0
        float_frequency = reset
        remainder = reset
        # print 'Normalized celsius: ', celsius
        if celsius == 0.0:
            float_frequency = floor_frequency    
        else: 
            float_frequency = (math.pow(octave_ratio, celsius) * floor_frequency)
            # # print 'Pow... ', math.pow(octave_ratio, celsius) # debug
        # print 'Frequency: ', float_frequency
        int_frequency = int(float_frequency)
        # print 'Int Frequency: ', int_frequency
        remainder = (float_frequency%int_frequency)
        pitch_bend_val = int(remainder * maxpitchbend) 
        # print 'Remainder: ', remainder
        # print 'PitchBend Val: ', pitch_bend_val
        note_num = pretty_midi.hz_to_note_number(int_frequency)
        note_num = int(note_num)
        # print 'Midi note number = ', note_num
        # print 'Time = ', time
        note = pretty_midi.Note(velocity = velocity_value, pitch = note_num, start = time, end = time + time_shift)
        bend_val = pretty_midi.PitchBend(pitch_bend_val, time)
        time = time + time_shift
        pretty_midi_instrument.notes.append(note)
        pretty_midi_instrument.pitch_bends.append(bend_val)
        if percussion_flag == True:
            # print 'Writing to percussion track' # debug
            percussion.notes.append(note)
    print ''
    
def convert_diff_data_to_midi(diff_array, pretty_midi_instrument):
    
    time = reset 
    
    for celsius in diff_array:
            
        if celsius == 0.0:
            celsius = global_diff_min
        # print 'Celsius:', celsius 
        celsius = normalize(celsius, global_diff_min, global_diff_max) # convet to 0.0 - 1.0, normalized only for diff functions
        float_frequency = reset
        remainder = reset
        sign = 1
        # Rule for calculating ROC:
        if(celsius < 0): # if negative:
            sign = -1
            celsius = abs(celsius)
        else: # Redundant test debug
            sign = 1 # Redundant test debug
        # print 'Normalized celsius: ', celsius
        if celsius == 0.0:
            # print("We got a zero..")
            float_frequency = 1   
        elif sign == 1: 
            float_frequency = (math.pow(octave_ratio, celsius) * floor_frequency)
            ## print 'Pos Pow... ', math.pow(octave_ratio, celsius)
            # print 'POS'
        elif sign == -1:
            #float_frequency = (ceiling_frequency/(math.pow(octave_ratio,celsius))) #divide? test debug
            float_frequency = (math.pow(octave_ratio, celsius) * floor_frequency) # test debug
            # print 'NEG'
        # print 'Frequency: ', float_frequency
        int_frequency = int(float_frequency)
        # print 'Int Frequency: ', int_frequency
        if (float_frequency != 1): #skipping no changerem
            remainder = (float_frequency%int_frequency)
            pitch_bend_val = int(remainder * maxpitchbend)
            # print 'Remainder: ', remainder
            # print 'PitchBend Val: ', pitch_bend_val
            note_num = pretty_midi.hz_to_note_number(int_frequency)
            note_num = int(note_num)
            # print 'Midi note number = ', note_num
            # print 'Time = ', time
            note = pretty_midi.Note(velocity = velocity_value, pitch = note_num, start = time, end = time + time_shift)
            bend_val = pretty_midi.PitchBend(pitch_bend_val, time)
        time = time + time_shift
        pretty_midi_instrument.notes.append(note)
        pretty_midi_instrument.pitch_bends.append(bend_val)
    print ''
    
#                                                                                 #
###################################################################################
floor_frequency = 55 #440 # 'A4', MIDI note 69
ceiling_frequency = 220
convert_data_to_midi(new_data_core_csv, cello)
floor_frequency = 220 #440 # 'A4', MIDI note 69
ceiling_frequency = 440
convert_data_to_midi(new_data_distal_csv, piano)
floor_frequency =  440 #440 # 'A4', MIDI note 69
ceiling_frequency = 1760 
convert_data_to_midi(new_data_axial_csv, synth)

# print('\n\n\n\n\n')
convert_diff_data_to_midi(new_diff_core_csv, flute)
convert_diff_data_to_midi(new_diff_distal_csv, clarinet)
convert_diff_data_to_midi(new_diff_axial_csv, piccolo)


# max...

# Add the instruments to the PrettyMIDI object

temperature_sonify.instruments.append(cello) # core
temperature_sonify.instruments.append(piano) # distal
temperature_sonify.instruments.append(synth) # axial
temperature_sonify.instruments.append(flute) # diff_core
temperature_sonify.instruments.append(clarinet) # diff_distal
temperature_sonify.instruments.append(piccolo) # diff_axial
temperature_sonify.instruments.append(percussion) #events

# Write out the MIDI data
temperature_sonify.write('SonifiedTemperatureData.mid')

```
